## Six months is more than enough time to make a difference

Anyway, I wanted to use this post to reflect on the value I've created for Sourceress over the last half year or so. The conventional wisdom is that you typically leave these shorter positions off of your resume because they're a red flag to potential employers, or because you need a few months to be truly effective, but look at how I hit the ground running in the first couple months:

- Dev-ops:  automating and hardening AWS deployments, particularly with Bash and Python scripts for CloudFormation and Elastic Beanstalk, and setting up critical configurations and permissions.
- Back- and front-end:  creating training modules for the internal team such that new team members can be onboarded more effectively; training cases can be created, edited, and reviewed by current team members, accessed and answered by new team members.
- Front-end:  creating a highlighting feature very similar to the one [found at Medium](https://help.medium.com/hc/en-us/articles/214406358-Highlighting), meant to improve team member efficiency, speed knowledge transfer, and gather text for internal NLP models.
- Design:  improve the UI of the internal site to make the team more efficient and effective:  cleaning and standardizing the presentation of raw text content, strategically ordering and sizing content based on the value of that information, removing content that only serves to distract or clutter, generating charts to impart quantitative information more quickly.

And hey, did I mention that was all within the first two months? Why wouldn't I want to emphasize the impact of my work, especially when I learned how to work with things like Elastic Beanstalk and Handlebars.js on the fly? This is exactly the type of full-stack, can-do experience needed by small teams.

## Calculating hard numbers on value-added

I'll admit that the value of the above work is relatively vague. I could dig up internal numbers from this time period to try and quantify it's value, but it would be impossible to separate the value of my work from the concurrent features and improvements made by Josh, our co-founder and only other engineer at the time. However, I was the only individual working on our foundational machine learning infrastructure, meaning I can do some simple back-of-the-envelope calculation to estimate the value of our models.

### Business overview

First, what's the high-level overview of our process?
- We have team members called "raters" who review candidate profiles to find people who are great fits for our partners.
- Each candidate has a short profile and a long profile; if a candidate's short profile looks like a potential fit, a rater will also review a long profile.
- The time to review short profiles is ~30 seconds and 20\% are fits.
- The time to review long profiles is ~2 minutes and 10\% are fits.

### Value-added to the short profile workflow

Originally, raters would get a batch of profiles, review all profiles, and move to the next batch. Given the parameters above and assuming batches of 1,000 short profiles, half of the approved candidates are found after **4.16 hours**.

To make this more efficient, I ranked candidates according to their likelihood of being approved. Specifically, I created a data pipeline and an ensemble of machine learning models, which included natural language processing. Luckily, I also created a custom analytics dashboard so that I can quickly quantify model performance.

Now, half of the approved candidates are found after only **1.04 hours**

In other words, raters can find **4 times as many great candidates** in the same amount of time.

### Value-added to the long profile workflow

Given the parameters above and assuming batches of 1,000 long profiles, half the approved candidates are found after **16.66 hours**.

Now, given my profile ranking, half of the approved candidates are found after only **2.5 hours**.

In other words, raters can find **6.66 times as many great candidates** in the same amount of time.

### The bigger picture

Without giving financial details and with the caveat that it's too early to see the full effect of these improvements, I estimate that my improvements to our rating efficiency will lead to a **4-10x increase in revenue**. How did I get this number? I made reasonable assumptions using the time saved by raters and the quality of candidates after being sorted.

Too wishy-washy? The raters' salaries account for **~20% of our monthly budget**. The machine learning infrastructure I've created is going to save Sourceress approximately **~20% of that value** or **~4% of our total budget**. That might seem insignificant, but it actually means we've extended our runway by a fair amount, with me keeping that number close to the chest because it would allow one to calculate potentially sensitive numbers.

## How do you remember all of this?

I don't. I'm doing too many different things too quickly to remember what's happened even a few weeks ago, which is why I'm meticulous about code cleanliness, object and variable names, and documentation and comments.

Instead, I keep a log where I summarize the tasks I've completed, the new technologies or skills I've learned, and the features I've shipped every week. It's incredibly valuable for tracking my progress and understanding my strengths (so that I can sell myself in interviews) and weaknesses (so that I can actually improve professionally and personally -- I'd argue this is more important). 

I'd recommend keeping your own log and updating it every Friday afternoon.
