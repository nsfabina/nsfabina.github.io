# Thoughts from my recent job search... again

I wanted to reflect on my most recent job search, having recently joined ERA Economics as CTO. Believe it or not, I've received a fair amount of positive feedback on my <a href="http://nbviewer.jupyter.org/github/nsfabina/nsfabina.github.io/blob/master/ipynb/job_search.ipynb" target="_blank">previous job search post</a>, and I found it useful to adjust my recent search strategy after revisiting that content. As with before, these reflections are based on my own personal and professional goals, and alternative priorities and viewpoints are completely reasonable.

## Strong signals

I'm particularly interested in the gradient between work-life balance and work-life integration, with my personal preferences falling closer to work-life balance. That is, I'm a great fit for organizations where team members are productive and enthusiastic about a shared mission, and balance professional goals with personal considerations like family, friends, and outside interests.

How do I identify places with a culture of work-life balance? I look for websites or job posting where organizations talk about their team members' hobbies, quirks, and especially parenting obligations. In some cases, organizations will come right out and describe how they build a culture that supports work-life balance.

It's slightly more difficult to identify places with work-life integration, perhaps because candidates would be more likely to filter out these organizations. Instead, I look for certain euphemisms, and dig deeper during initial phone screens. The following phrases are relatively informative signals:
- "We write code quickly and deploy frequently"
- "We're looking for mission-driven engineers who want to change the world"
- "We're hiring a (rockstar|ninja|wizard) who can be a 10x contributor"

To be clear, I'm not saying balance is better than integration, but that assessing culture fit is critical. For example, one of my former colleagues fondly recalls a previous position where team members ate breakfast, lunch, and dinner together, talked shop until late in the evening, and met up to hack with each other on the weekend. As candidates, we are both responsible for identifying whether potential employers have a culture that matches with our personal preferences, because a mismatch is going to lead to dissatisfaction for employee, employer, or both!

## Instant disqualifier

During my last search, I was more willing to do coding tests because I had a smaller professional network and my technical experience was less visible. Unfortunately, the companies that would assign coding tests prior to any human interaction were also the companies that were less likely to give any response.

This time around, I would send polite responses that highlighted my publicly available code samples and ask that organizations contact me if they're interested in particular aspects of my experience. Organizations that responded favorably were likely to be enthusiastic about me joining their team and often great fits. Organizations that gave negative or no response were filtering themselves out by not being sufficiently interested in me as a candidate or by not valuing my time and energy. These places were also more likely to be collecting resumes but not actively hiring, which is not a good time investment on my end.

## Evaluating compensation

Let's pretend you have two offers:

Offer 1 is a stereotypical compensation package from a well-funded, Bay Area startup:
- $150k
- unlimited vacation
- no retirement plan
- work-life integration, or approximately 12hrs/day on average

Offer 2 is a stereotypical compensation package from a small, established business:
- $100k
- three weeks vacation
- 401k with no match
- work-life balance, or approximately 8hrs/day on average

(These are obviously generalizations, as many startups give well-rounded compensation packages too, but let's stick with the stereotypes for illustration.)

All else equal, how do you compare these offers? It might make a lot of sense to take the first offer if you're trying to optimize salary.

What happens when we convert yearly salary to dollars per hour, rather than dollars per year?
- Offer 1:  \$48/hr 
- Offer 2:  \$48/hr 

What about after-tax compensation?
- Offer 1:  \$36/hr
- Offer 2:  \$37/hr

What if we assume that unlimited vacation often means "no vacation" in practice?
- Offer 1:  \$36/hr
- Offer 2:  \$41/hr

What if we're interested in saving for the future and we max out our 401k vs. using non tax-advantaged accounts?
- Offer 1:  \$36/hr
- Offer 2:  \$43/hr

Again, all else equal, we maximize our hourly wage by taking the second position. Even if we want to maximize our yearly salary and don't care about inefficiency, we're still better off finding a side-gig such that the combination of offer 2 and the side-gig have a greater expected value than offer 1.

Just for fun, what about a third, stereotyped offer, this time from a poorly-funded startup, perhaps where founders are trying to extend their runway? Let's assume this offer is otherwise similar to the first offer, but with a salary in line with the second.

Offer 3:
- $100k
- unlimited vacation
- no retirement plan
- work-life integration, or approximately 12hrs/day on average

After taxes, this is an hourly wage of $25/hr -- almost half the second offer!

Sure, there may be personal preferences or professional development reasons to choose a lower hourly rate, or positions with a lower hourly rate may offer significantly more equity, but it makes sense to dig deeper when choosing between positions.

## Titles matter

My first industry position was as a data scientist at FBN, but I had significant software engineering responsibilities. I highlighted this experience during my previous job search and had to convince potential employers that they should take my background seriously.

My second industry position was as a software engineer at Sourceress, but I developed a fair number of data science assets during my tenure. During my most recent search, potential employers seemed to take my software engineer experience at face value, and I felt as if interviewers were less likely to ask pointed questions about my skill set.
