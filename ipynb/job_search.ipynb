# <a name="requirements">Position requirements</a>

I had the following requirements for potential positions:
- I wanted a data science or software engineering position, although I was flexible on the exact responsibilities as long as I could continue learning new things
- I wanted a remote position for the ability to support my partner's career or relocate as desired
- I wanted an atmosphere that was positive, collaborative, supportive, etc - I wanted wonderful colleagues and shared goals
- Bonus points for achieving meaningful social outcomes

My requirements changed slightly over the course of the search. I was reconnecting with old friends, playing soccer and basketball, practicing yoga and competing at pub trivia, and just generally enjoying myself. I felt that a remote position which allowed me to maintain that life balance would be more desireable than a position that changed the world but restricted my free time.

# <a name="constraints">Search constraints</a>

I had the following constraints to consider during my search:
- my remote requirements immediately restricted the pool of potential positions, especially with respect to data science
- I had to apply rather than ask for internal references, as my current professional network has very little overlap with the tech community more generally
- I had to take more coding tests and jump through more hoops than I would have preferred because my professional experience and public code samples were limited
- I refuse to study for job interviews or coding tests because I believe I should be hired based on a lifetime of experience rather than a piece of trivia, and I occasionally performed poorly when interviewed on the basis of algorithmic trivia

# <a name="process">Application process</a>

I gradually settled on the following process for finding positions:
1. Search {position resource} for all remote positions
2. Filter for positions that are interesting
3. Filter for positions whose requirements I satisfy
4. Repeat until I have a list of ~10-15 high value opportunities

I would then apply to positions until my brain just couldn't handle it any longer. In practice, this meant slightly modifying one of about five cover letter templates for a specific position, and sending an email with that cover letter and my CV. On average, I was applying to three organizations every weekday and taking approximately a week to search for and apply to a full list of positions.

At the time that I accepted my new position:
- 26 organizations were applied to
- 3 never responded
- 4 rejected immediately
- 3 rejected after coding test
- 4 rejected after interviews
- 4 never responded after interviews
- 4 wanted to continue the interview process, but were moving slowly
- 4 were in the process of making or had made an offer

All considered, 8 applications had positive outcomes (31%) and 18 had negative outcomes (69%). I feel that there was a strong element of luck involved because I did not notice any strong patterns. Surprisingly, I would be rejected by social good organizations which should have been excited about my experience and motivations, while other companies would interview me for front-end positions despite my limited experience.

# <a name="work">Hiring practices that work well</a>

The following hiring practices are pretty fantastic because they expand the pool of qualified candidates that apply and they give applicants a positive impression of the organization:
- explicitly mentioning that number of years experience, in total or with specific languages or tools, is not required
- describing the full application process, tracking progress through the system, and estimating timelines
- asking about previous projects and experiences and adapting interviews to fit the candidate; e.g., waiving rudimentary coding challenges when code samples are already available
- offering the option for skipping coding challenges for small, paid sample projects
- giving specific and meaningful feedback on interview performance

# <a name="dontwork">Hiring practices that don't work so well</a>

The following hiring practices are pretty terrible, leave a candidate with a poor impression of an organization, and can be improved pretty easily:

Practice:  no response after application received.  
Alternative:  send an automated, impersonal response to all candidates that are rejected.  

Practice:  no response after interviewing the candidate.  
Alternative:  send a semi-automated, somewhat personalized response to any candidate that has had at least one round of interviews.  

Practice:  asking all applicants to finish coding challenge before a human looks at application materials.  
Alternative:  have someone do a very basic filter on applications before asking applicants to spend an hour taking a coding test.  

Practice:  rejecting a candidate for a reason that was not included on the job posting.  
Alternative:  review your posting carefully and/or solicit outside input to ensure that applicants know your requirements, can filter themselves out if unqualified, or can highlight their strengths appropriately.  

Practice:  not hiring for a position after interviewing candidates in earnest.  
Alternative:  do not interview candidates if you're not sure of your organization's goals, the position's responsibilities, or what a successful candidate looks like.  
		
# <a name="paidoff">Things that paid off</a>

The three things that were most useful during my search:
1. I was very good at highlighting the professional and independent nature of my PhD, which meant that organizations could count my years of academic research as relevant work experience. I could demonstrate that I had ten years of relevant data science and engineering experience, rather than allowing others to view me as a fresh PhD with only two years in industry.
2. I had learned new skills and generated project samples by working on side projects and reading broadly during evenings and weekends. In fact, I'm primarily working in Django at my new position, and I learned how to use Django while working on my own side project over several weekends.
3. I created a compelling narrative with my background experience that caught the attention of organizations and colleagues with shared values. I wasn't a data scientist with extensive statistics and machine learning skills; I was a person who had worked to save coral reefs, protect endangered species, and grow crops more efficiently, with the help of quantitative and computational tools.

I note that all three require long-term investments. The first requires practicing effective communication skills and self-promotion, the second involves developing habits to continue learning, and the third is finding ways to have your career reflect your values.
